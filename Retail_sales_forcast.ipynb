```python
#Hypotheses

continues=['Temperature', 'Fuel_Price', 'MarkDown1', 'MarkDown2',
       'MarkDown3', 'MarkDown4', 'MarkDown5', 'CPI', 'Unemployment',
       'Weekly_Sales','Size', 'Week_of_year']
categories=['Store', 'IsHoliday', 'Dept', 'Type']


from scipy import stats

def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.055:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False

#defining function for categories vs categories
def chisqare_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.055 else False

def annova_test(d1,d2):
  group=df_final[d2].unique()
  data={}
  for i in group:
    data[i]=df_final[d1][df_final[d2]==i]
  f_value,p_value=stats.f_oneway(*[i for i in data.values()])
  if p_value < 0.055:
    return False
  else:
    return True

final={}
for i in df_final.columns:
  final[i]={}
  for j in df_final.columns:
    if (i in continues) and (j in continues):
      result=two_sample(df_final[i],df_final[j])
    elif  (i in continues) and (j in categories):
      result=annova_test(i,j)
    elif (i in categories) and (j in continues):
      result=annova_test(j,i)
    elif (i in categories) and (j in categories):
      result=chisqare_cat_vs_cat(df_final[i],df_final[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x=df_final.drop("Week_of_year",axis=1)
y=df_final["Week_of_year"]
```

```python
from tensorflow.keras.utils import to_categorical
ynew=to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,ynew,test_size=0.3)
```

```python
x_train_scaler=scaler.fit_transform(x_train)
```

```python
x_test_scaler=scaler.transform(x_test)
```

```python
x_train_scaler.shape[1]
```

```python
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential

import tensorflow as tf
model=Sequential()
model.add(Dense(78125,input_dim=x_train_scaler.shape[1],activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.2))
model.add(Dense(15625,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.2))
model.add(Dense(3125,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.23))
model.add(Dense(625,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.15))
model.add(Dense(125,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dense(25,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dense(1))
model.compile(loss="mean_squared_error",optimizer="adam",metrics=[tf.keras.metrics.R2Score(),tf.keras.metrics.MeanAbsoluteError(),tf.keras.metrics.MeanSquaredError(),tf.keras.metrics.MeanAbsolutePercentageError()])
```

```python
result=model.fit(x_train_scaler,y_train,epochs=30,validation_data=(x_test_scaler,y_test),batch_size=15)
```
